```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

data=pd.read_csv(r"D:\5 th sem\ml\headbrain.csv")
x=data["Head Size(cm^3)"]
y=data["Brain Weight(grams)"]

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)

model=LinearRegression()
model.fit(np.expand_dims(x_train,axis=1),np.expand_dims(y_train,axis=1))
pred=model.predict(np.expand_dims(x_test,axis=1))

plt.figure(figsize=(10,5))
plt.scatter(x_test,y_test,c="r",label="Scatter line")
plt.plot(x_test,pred,c="b",label="Regresion")
plt.title("head Size vs Brain Weight")
plt.xlabel('head size')
plt.ylabel('brain weight')
plt.legend()
plt.show()

```

```python
r=model.score(np.expand_dims(x_test,axis=1),np.expand_dims(y_test,axis=1))
mse=mean_squared_error(y_test,pred)
print(f"rmse{np.sqrt(mse)}\n r2 Score{r}")
```

# Univariate

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

x=np.linspace(-3,5,150)
params=[(0,1,"$N(0,1)$"),(1,2,"$N(1,2)$"),(1,0.2,"$N(1,0.2)$")]

fig,ax=plt.subplots(figsize=(10,5))
for mean,cov,label in params:
    pdf=norm.pdf(x,loc=mean,scale=np.sqrt(cov))
    ax.plot(x,pdf,label=label)

plt.title("Univariate")
plt.xlabel('$x$ ')
plt.ylabel('Density:$p(x)$')
plt.legend()
plt.show()
```

# Multivarite 

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
def multi(mean,cov):
    x,y=np.linspace(-5,5,100),np.linspace(-5,5,100)
    x1,x2=np.meshgrid(x,y)
    pos=np.dstack((x1,x2))
    pdf=multivariate_normal(mean,cov).pdf(pos)
    return x1,x2,pdf

d=[(np.array([0,0]),np.eye(2)),(np.array([0,0]),np.array([[1,-0.1],[-0.8,1]]))]

fig,ax=plt.subplots(1,2,figsize=(10,3))
plt.suptitle("Bivariate")
ax[0].set_title('Independent')
ax[1].set_title("covariance")

for i,dist in enumerate(d):
    x1,x2,pdf=multi(dist[0],dist[1])
    con=ax[i].contourf(x1,x2,pdf,100,cmap='rainbow')
    ax[i].set_xlabel('$x_1$',fontsize=13)
    ax[i].set_ylabel('$x_2$',fontsize=13)
    ax[i].axis([-3.5,3.5,-3.5,3.5])
    c_bar=fig.colorbar(con,ax=ax[i])
    c_bar.ax.set_ylabel('$y(x1,x2)$',fontsize=13)

plt.show()
```

# Kmeans Cluster

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans as km
from sklearn.metrics import silhouette_score

data=pd.read_csv(r"D:\5 th sem\ml\iris.csv")
x=data[["sepal_length","petal_length"]]

def kmeans(x,k):
    d=[]
    for i in range(1,k+1):
        model=km(n_clusters=i,init='random',n_init='auto',max_iter=10).fit(x.values)
        d.append(model.inertia_)
    print(f"sil{silhouette_score(x,model.labels_)}")
    return model,d

k=5
model,d=kmeans(x,k)

```

```python
cen=model.cluster_centers_
index=model.labels_

plt.figure(figsize=(10,5))
plt.scatter(cen[:,0],cen[:,1],c="b",marker="+",linewidth=16)
sns.scatterplot(data=data,x="sepal_length",y="petal_length",hue=index,palette="coolwarm")

plt.title("Cluster")
plt.xlabel("sepal")
plt.ylabel("petal")
plt.grid()
plt.savefig("kmeans")
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(x=np.arange(1,k+1),y=np.array(d),marker="o",markersize=10)
plt.yticks(np.arange(1,k+1))
plt.title("Elbow method of optimal k")
plt.xlabel("no of clusters")
plt.ylabel("loss")
plt.show()
```

# GMM

```python
import numpy as np
from sklearn.mixture import GaussianMixture
from matplotlib.patches import Ellipse 

def gmm(k=3,dim=2,p=500,lim=[-10,10]):
    x=[]
    mean=np.random.rand(k,dim)*(lim[1]-lim[0])+lim[0]
    for i in range(k):
        cov=np.random.rand(dim,dim+10)
        cov=np.matmul(cov,cov.T)
        x_=np.random.multivariate_normal(mean[i],cov,p)
        x+=list(x_)
    x=np.array(x)
    if (dim==2):
        fig=plt.figure(figsize=(10,5))
        ax=fig.gca()
        ax.scatter(x[:,0],x[:,1],s=40,alpha=0.7)
    return x

data=gmm()
model=GaussianMixture(n_components=3)
model.fit(data)

fig=plt.figure(figsize=(10,5))
plt.scatter(data[:,0],data[:,1],s=40)

for mean,cov,c in zip(model.means_,model.covariances_,["r","y","k"]):
    v,w=np.linalg.eigh(cov)
    angle=np.arctan2(w[0][1],w[0][0])
    angle=angle*180/np.pi
    v=2.0*np.sqrt(2.0)*np.sqrt(v)
    ell=Ellipse(mean,v[0],v[1],angle=180+angle,alpha=0.4,color=c)
    plt.gca().add_patch(ell)

plt.title("gmm")
plt.xlabel("no of clusters")
plt.ylabel("loss")
plt.show()  
```

```python


def plotq(model,x,y):
    x_min,x_max=x[:,0].min()-0.1,x[:,0].max()+0.1
    y_min,y_max=x[:,1].min()-0.1,x[:,1].max()+0.1

    xx,yy=np.meshgrid(np.linspace(x_min,x_max,100),np.linspace(y_min,y_max,100))

    x_=np.c_[xx.ravel(),yy.ravel()]
    y_pred=model.predict(x_)

    if len(y_pred)>1:
        print("doing Multiclass classifiaction")
        y_pred=np.argmax(y_pred,axis=1).reshape(xx.shape)
    else:
        print("doing Binary")
        y_pred=np.round(y_pred).reshape(xx.shape)

    plt.contourf(xx,yy,y_pred,cmap=plt.cm.RdYlBu,alpha=0.3)
    plt.scatter(x[:,0],x[:,1],c=y,s=40,cmap=plt.cm.RdYlBu)
    plt.xlim(xx.min(),xx.max())
    plt.ylim(yy.min(),yy.max())

plt.subplot(1,2,1)
plt.title("Training")
plotq(model,x_train,y_train)
plt.subplot(1,2,2)

plt.title("testing")
plotq(model,x_test,y_test)
plt.show()
```

# CNN

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import (
Conv2D, MaxPooling2D,
Dense, Dropout, Flatten,
BatchNormalization
)
from tensorflow.keras.utils import to_categorical
import joblib

train =pd.read_csv(r"D:\5 th sem\ml\mnist_train.csv")
test =pd.read_csv(r"D:\5 th sem\ml\mnist_test.csv")
X_train = train.drop(['label'], axis=1).values
y_train = train['label'].values
X_test = test.values
X_train = X_train.reshape(-1,28,28,1).astype('float32')/255
X_test = X_test.reshape(-1,28,28,1)
X_train = X_train /255
X_test=X_test/255
y_train=to_categorical(y_train)

fig, axes = plt.subplots(3, 3, figsize=(6, 6))
for i, ax in enumerate(axes.flat):
    ax.imshow(X_train[i].squeeze(), cmap='binary')
    digit = y_train[i].argmax()
    ax.set(title = f"label: {digit}")
fig.tight_layout()

def get_model():
    model=Sequential()
    model.add(Conv2D(filters=64, kernel_size = (3,3), activation="relu",
    input_shape=(28,28,1)))
    model.add(Conv2D(filters=64, kernel_size = (3,3), activation="relu"))
    model.add(MaxPooling2D(pool_size=(2,2)))
    model.add(BatchNormalization())
    model.add(Conv2D(filters=128, kernel_size = (3,3), activation="relu"))
    model.add(Conv2D(filters=128, kernel_size = (3,3), activation="relu"))
    model.add(MaxPooling2D(pool_size=(2,2)))
    model.add(BatchNormalization())
    model.add(Conv2D(filters=256, kernel_size = (3,3), activation="relu"))
    model.add(MaxPooling2D(pool_size=(2,2)))
    model.add(Flatten())
    model.add(BatchNormalization())
    model.add(Dense(512,activation="relu"))
    model.add(Dense(10,activation="softmax"))
    model.compile(loss="categorical_crossentropy", optimizer="adam",
    metrics=["accuracy"])
    return model


tf.random.set_seed(0)
model = get_model()
print(model.summary())


history = model.fit(
X_train, y_train,
batch_size = 64,
epochs = 20,
validation_split=.2
)
tf.keras.models.save_model(model,"models/mnist_cnn.h5")
joblib.dump(history.history, "models/mnist_cnn.history")
model = tf.keras.models.load_model("models/mnist_cnn.h5")
history = joblib.load("models/mnist_cnn.history")

fig, ax = plt.subplots(1,2, figsize=(15,5))
ax[0].plot(history['loss'], color='b', label="Training loss")
ax[0].plot(history['val_loss'], color='r', label="validation loss",axes =ax[0])
ax[0].legend()
ax[1].plot(history['accuracy'], color='b', label="Training accuracy")
ax[1].plot(history['val_accuracy'], color='r',label="Validation accuracy")
ax[1].legend()

y_pred = model.predict(X_test)
fig, axis = plt.subplots(4, 4, figsize=(8,10))
for i, ax in enumerate(axis.flat):
    ax.imshow(X_test[i].squeeze(), cmap='binary')
    ax.set(title = f"Prediction: {y_pred[i].argmax()}");
    ax.axis("off")
    fig.suptitle("Test Predictions")
fig.tight_layout(rect=[0, 0.05, 1, 0.95])
```
